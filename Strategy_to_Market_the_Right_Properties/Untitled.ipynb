```python
grouped_by = accounts_opportunities_properties.groupby('id_accs').groups
large_customers=[]
small_customers=[]


for cust,count in grouped_by.items():
    if len(count) >2:
        large_customers.append(cust)
    else:
        small_customers.append(cust)
print("large",len(large_customers))
print("small",len(small_customers))
```
